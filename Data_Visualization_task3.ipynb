### Data Visualization (Task 3) Script

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter # Useful for counting actor occurrences

# --- 1. SETUP: LOAD CLEANED DATA ---
cleaned_csv_file = 'imdb_top_25_cleaned_data.csv'
print(f"--- Phase 3: Starting Data Visualization on {cleaned_csv_file} ---")

try:
    df = pd.read_csv(cleaned_csv_file)
    print("✅ Cleaned data loaded successfully.")
except FileNotFoundError:
    print(f"❌ ERROR: Cleaned file '{cleaned_csv_file}' not found. Please run Phase 2 (EDA) first.")
    exit()

# Configure visualization styles
sns.set_style("darkgrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12
```

```python
# --- 2. DATA TRANSFORMATION FOR ACTOR ANALYSIS ---

# Rationale: The 'Main Characters' column is a single string of comma-separated names. 
# We need to split, flatten, and count these names to find the most frequent actors.
def count_actors(df):
    """Splits and counts the frequency of all actors in the 'Main Characters' column."""
    all_actors = []
    # Iterate through the Main Characters column
    for cast_string in df['Main Characters']:
        if pd.notna(cast_string) and cast_string != 'N/A':
            # Split the string by comma and remove leading/trailing spaces
            actors = [actor.strip() for actor in cast_string.split(',')]
            all_actors.extend(actors)
    
    # Count the frequency of each actor
    actor_counts = Counter(all_actors)
    # Convert to DataFrame for easy plotting and select top 10
    top_10_actors = pd.DataFrame(actor_counts.most_common(10), columns=['Actor', 'Count'])
    return top_10_actors

top_actors_df = count_actors(df)
print(f"✅ Actor frequency analysis completed. Found {len(top_actors_df)} unique top actors.")
```

```python
# --- 3. GENERATE VISUALS (Task 3 Requirements) ---

# Visual 1: Top 10 Most Frequent Actors (Bar Plot)
plt.figure(figsize=(12, 7))
sns.barplot(x='Count', y='Actor', data=top_actors_df, palette='Spectral')
plt.title('Top 10 Most Frequent Actors in IMDb Top 25 Movies', fontsize=16)
plt.xlabel('Number of Appearances', fontsize=14)
plt.ylabel('Actor', fontsize=14)
plt.show()


# Visual 2: Rating Distribution (Box Plot)
# Rationale: A Box Plot is excellent for visualizing the spread, median, and identifying outliers.
plt.figure(figsize=(8, 6))
sns.boxplot(y=df['Rating'], color='lightblue')
plt.title('Rating Distribution (Box Plot)', fontsize=16)
plt.ylabel('IMDb Rating', fontsize=14)
plt.show()


# Visual 3: Rating vs. Year (Scatter Plot - Enhanced)
# This chart was introduced in EDA, but is crucial for visualization and insight communication.
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Year', 
    y='Rating', 
    data=df, 
    s=200, # Increased size for impact
    hue='Rating', 
    palette='magma', 
    alpha=0.8,
    edgecolor='black'
)
correlation = df['Year'].corr(df['Rating'])
plt.title(f'Movie Release Year vs. IMDb Rating (Correlation: {correlation:.2f})', fontsize=16)
plt.xlabel('Release Year', fontsize=14)
plt.ylabel('IMDb Rating', fontsize=14)
plt.legend(title='Rating')
plt.show()


print("\n🎉 Phase 3: Data Visualization Completed!")
print("Three key visualizations have been generated and displayed.")
```
